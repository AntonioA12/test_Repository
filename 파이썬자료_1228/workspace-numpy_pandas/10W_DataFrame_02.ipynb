## DataFrame 의 요소 

```python
dic1= {'a':[1,2,3],'b':[2,3,4], 'c':[3,4,5]}

df1 = pd.DataFrame(dic1 , index=['g','k','p'])

df1

df1.index.name='inds'
df1.columns=list('uio')
df1 = df1.astype(float)
df1


```

## 특정 columns 데이터타입 , 데이터  변경  / 묶음 변경

```python
df1['u'] = df1['u'].astype(int) 
df1['i'] = [10,20,30]

df1

# [:]  을 사용에 특정 행&열 의 데이터를 묶음 변경 가능 

df1[:] = np.arange(100,109).reshape(3,3)
df1
```

## 데이터타입이 중요한 이유 1 

    - DataFrame 은 object를 다루기 떄문에
    - Column 안의 있는 Data 가 int 와 float 아니더라도 str 과 연계하여 수학계산함수가 적용이 된다.
    - 단, str을 숫자가 아닌 str 자체로 계산하기 때문에 상식불가의 계산수가 나온다.
    
    - 물론 str 이 '10' , '20' 같이 숫자의 형태일 경우에만 해당
   
    - 특정 column 이나 특정 index 가 int , float 이 아니고 str 일경우 dtype 이 object 나온다. 
    - 이럴 경우 오타일 확률이 크다.

```python
dic2= {'a':[1,2,3],'b':[2,3,4], 'c':[30,40,50]}
df2 = pd.DataFrame(dic2 , index=['g','k','p'])

df2
df2.mean(axis=0)
df2['c']=df2['c'].astype(str)
df2
df2.mean(axis=0)
```

## 실습 2
    
    - DataFrame 을 생성하고 
    - DataFrame 의 인덱스, column , 데이터 를 변경해보자 

```python
dic3= {'a':[1,2,3],'b':[2,3,4], 'c':[3,4,5]}
df3 = pd.DataFrame(dic3 , index=['g','k','p'])

df3
df3.index =['aa','bb','cc']
df3.index.name='index'
df3.columns.name='column'
df3['b']=[11,22,33]
df3['b'] = df3['b'].astype(float)
df3
```
