## 종속변수 wage 에 대한 독립변수 OLS 계산 코멘트

    -  P-value 는 분석이나 예측하는데 있어서 매우 집중적으로 관찰해야할 값이다.
    - 그러나 종속변수들의 OLS 표를 보면 일부를 제외하곤 P-value 가 매우 높게 나오며, 
    - 이는 유의도가 매우 낮음을 알린다. ( 변수 변경요망 이 있음을 알려준다. ) 
    
    - 고맙게도 OLS 표 아래의  Warnnigs [2] 를 보면 
    - 변수의 수와 수치가 너무 커서 다중 공산성의 문제가 있음을 알려준다.

```python
reg1 = smf.ols( '''wage ~ hsiblings + hfathereduc + hmothereduc + siblings +
       lfp + hours + kidsl6 + kids618 + age + educ + federaltax  + wage76 +
       hhours + hage + heduc + hwage + faminc + mtr + mothereduc +
       fathereduc + unemployment + largecity + exper''', data=df  , missing='drop'   ).fit()
print(reg1.summary())
```

## VIF 계산식을 통해 다중공산성 의심되는 변수를 찾기

    - ViF factor를 10을 넘으면 다중공산성이 있다고 의심 
    
    - 만약 다중공산성의 위험이 있으면 Warnings :  에 코멘트가 추가 될것.
    - ex) [1] Standard Errors assume that the covariance matrix of the errors is correctly specified. = 이분산 문제가 존재한다.
    
#### 코멘트 & Question  

    - 아래 VIF 를 보면 P-value 가 안정적인 lfp , hours , edu , wage78 은 VIF 에서 높은 수치를 가졌고
    - 오히려 P-value 값이 지나치게 높은 변수들이 VIF 낮은 수치를 보여주고있습니다... 이유를 잘 모르겠습니다.

```python
features = '''wage + hsiblings + hfathereduc + hmothereduc + siblings +
       lfp + hours + kidsl6 + kids618 + age + educ + federaltax  + wage76 +
       hhours + hage + heduc + hwage + faminc + mtr + mothereduc +
       fathereduc + unemployment + largecity + exper'''

y , X = dmatrices('wage ~' + features , df , return_type='dataframe')

vif = pd.DataFrame()
vif['VIF Factors'] = [ variance_inflation_factor(X.values , i) for i in range(X.shape[1])]
vif['features']=X.columns

print(vif.round(1))
```

```python
results =sm.OLS( y , X).fit()
print(results.summary())
```

## 정규성 진단 - 정규분포의 형태가 나오는가?

### 히스토그램 
    - 아래의 'wage' 임금을 히스토그램 그래프로 보면 정규분포의 형태가 나오지않음을 알 수 있다.
### Q-Q plot
    - 파란색 점이 빨간선 의 형태와 유사하다면 정규성의 형태가 맞음을 알 수 있다.
### Jarque-Bera 
    - Jarque-Bera 값이 낮으면 낮을수록 정규분포의 형태를 가짐을 알 수있다.
    - twotail prob 의 수가 0에 가깔수록 유의도가 높음을 나타 낸다.

```python
# 히스토그램 
df['wage'].plot(kind='hist' , title='wage')
```

```python
# Q-Q plot
stats.probplot(reg1.resid , dist='norm' , plot=plt)
plt.title('Residual Q-Q plot')
```

```python
# jarque-Bera
name = ['Jarque-Bera','Chi^2 two-tail prob.' , 'Skew' , 'Kurtosis']
jarque = sms.jarque_bera(reg1.resid)
lzip( name , jarque)
```

## 정규성의 문제 해결법

```python
df['lnhwage'] = np.log(df['hwage']) 
```

```python
reg2 = smf.ols( '''lnwage ~ federaltax + hsiblings + hfathereduc + hmothereduc + siblings +
       lfp + hours + kidsl6 + kids618 + age + educ + wage76 +
       hhours + hage + heduc + hwage + faminc + mtr + mothereduc +
       fathereduc + unemployment + largecity + exper''' , missing='drop', data=df  ).fit()
print(reg2.summary())
```

```python
df['lnwage'].plot(kind='hist' , title='log of Wage') 
```

```python
stats.probplot(reg2.resid , dist='norm' , plot=plt)
plt.title('Residual Q-Q plot')
```

```python
name = ['Jarque-Bera','Chi^2 two-tail prob.' , 'Skew' , 'Kurtosis']
jarque = sms.jarque_bera(reg2.resid)
lzip( name , jarque)
```

```python
name = [ 'Lagrange_multiplier static','P-value','F-value','F_P-value']
BPtest = sms.het_breuschpagan(reg2.resid , reg2.model.exog)
lzip(name , BPtest)
```

```python
name = [ 'Lagrange_multiplier static','P-value','F-value','F_P-value']
WhiteTest = sms.het_white( reg2.resid , reg2.model.exog  ) # retres = False
lzip(name , WhiteTest)
```

```python
reg3 = smf.ols( '''lnhwage ~ federaltax + hsiblings + hfathereduc + hmothereduc + siblings +
       lfp + hours + kidsl6 + kids618 + age + educ + wage + wage76 +
       hhours + hage + heduc + hwage + faminc + mtr + mothereduc +
       fathereduc + unemployment + largecity + exper''' , missing='drop', data=df  ).fit(cov_types='HC3')
print(reg3.summary())
```
