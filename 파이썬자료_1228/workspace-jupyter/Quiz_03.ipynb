당신은 Ubar 서비스를 이용하는 개인 자가용 사업자입니다.

당신은 하루에 총 500분의 시간 동안 Ubar 서비스를 이용하여 승객을 매칭하고 이를 통해 수입을 얻습니다. 탑승 요금은 승객이 탑승하는 시간에 비례하여 받게 되며 1분에 500원을 받는다고 합니다. 탑승 고객은 다음의 조건을 만족하고, 탑승을 마친 후 다음 승객을 매칭하기까지 걸리는 시간은 10분이라고 가정합니다.

 

조건1: 승객별 운행 소요시간은 5분~55분 사이의 난수로 정해집니다.

조건2: 당신은 소요 시간 10분 ~ 30분 사이의 승객만 매칭해야 합니다.

 

이 경우, 당신은 하루에 Ubar 서비스를 이용하여 총 얼마의 수입을 얻게 되나요?

 

(출력문 예제)

당신은 오늘 하루 동안 총50명의 승객 중 20명의 승객과 매칭하여 10만원의 수익을 얻었습니다.

오늘 하루 수고하셨습니다.

```python
from random import *

count = 0 
money = 0 
w_time = 0 # 근무시간 500분 
m_time =10 # 매칭대기시간 10분 

for i in range (1,1000): # 승객 수  
    time = randrange(5,55) # 5분 ~ 50분 소요시간 인 승객 
    if 10<= time <=30: # 10분 ~ 30분 사이의승객만 픽업 
        count += 1 
        money += time * 500
        w_time += time # 총 근무시간 
        print(" {0}번째 손님 픽업! ( 소요시간 {1}분 , 토탈 근무시간 {2} 분 )".format(i , time ,w_time ))
        w_time += m_time # 매칭 시간 10분 
            
        if w_time + time >= 500: 
            print(' 근무시간 초과 ! 퇴근합시다.')
            break
            
    else : 
        print(" {0} 번째 손님 (소요시간 {1}분) 다른손님을 탐색합니다......".format(i , time))
        
print('=' * 100)        
print(' 당신은 오늘하루 동안 총 {0}명의 승객중 {1} 명의 승객과 매칭하여 {2} 원의 수익을 얻었습니다'.format(i, count,money))
print( ' 오늘 하루 수고하셨습니다.')

## P.S 일부 가정

# 가정 1. 일한시간(500분)은 손님을 태운시간+ 매칭대시간 10분 을 포함한다.
# 가정 2. 마지막손님은 500분이 초과해도 근무한다. 
```
