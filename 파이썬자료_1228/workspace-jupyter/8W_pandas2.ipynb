### 데이터 통합 ( 합치기) 

#### 세로 방향으로 통합
    -DataFrame_data1.append(추가할데이터 , [ , ignoer_index=True ] ) ignoer = 인덱스를 이어서 표시 한다면 T 
#### 가로 방향으로 통합
    -DataFrame_data1.join( 추가할데이터 ) 
    
#### 특정 열 기준으로 통합하기
    - DataFrame_left_data1(왼쪽).merge(DataFrame_right_data(오른쪽))

```python
df1 =pd.DataFrame({'class1':[95,92,98,100],
                  'class2':[91,93,97,99]})

df2 = pd.DataFrame({'class1':[87,89],
                   'class2':[85,90]})

df1
df2

df1.append(df2)
tdf = df1.append(df2, ignore_index=True)
```

#### 2행 자료가 부족해도 에러가 발생X 대신, nan 이 추가된다.

```python
df3 = pd.DataFrame({'class1':[96,92]})
df3

tdf
tdf.append(df3, ignore_index=True) # 2행 자료가 부족해도 에러가 발생X 대신, nan 이 추가된다.
```

```python
df4 = pd.DataFrame({'classe3':[93,91,95,100]})

tdf.join(df4)
```

```python
index_label = ['a','b','c','d']

dfla = pd.DataFrame({'class1':[95,92,98,100],
                  'class2':[91,93,97,99]},index=index_label)
df4a = pd.DataFrame({'classe3':[93,91,95,100]}, index=index_label)

dfla.join(df4a)
```

```python
df_A_B = pd.DataFrame({'판매월':['1월','2월','3월','4월'],
                      '제품 A':[100,150,200,180],
                      '제품 B':[90,120,150,160]})
df_C_D = pd.DataFrame({'판매월':['1월','2월','3월','4월'],
                      '제품 C':[120,130,210,150],
                      '제품 D':[70,90,110,130]})

df_A_B
df_C_D
df_A_B.merge(df_C_D)
```

#### .merge()

    - '판매월' 을 기준으로 merge
    - merge( data_name , how = left or right , on = ' ' )
    - merge( 데이터 이름, 어느방향으로 추가할건지 , 

```python
df_left = pd.pandas.DataFrame({'key':['A','B','C'],
                              'left':[1,2,3]})
df_right = pd.pandas.DataFrame({'key':['A','B','D'],
                              'right':[4,5,6]})

df_left
df_right

df_left.merge(df_right , how='left' , on = 'key')
```
