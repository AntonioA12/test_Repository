## add & drop column , row 
    - column 과 row 를 추가하거나, 제거하는 방법 
    
    - 주로 loc 를 사용, iloc 은 추가는 못하지만 변경은 가능하다 
    - element 갯수를 맞추기 않으면 당연히 에러가 발생 

```python
df1 = pd.DataFrame( np.arange(16).reshape(4,4) ,index=[10,20,30,40], columns=list('abcd'), dtype=float)

df1
df1.loc[50]=[16,17,18,19] # row 추가 
df1.loc[-1]=[10,20,30,40] # row 추가
df1
df1['e']=[1,2,3,4,5,6] # column 추가
df1

```

### 조건을 만들고, 조건을 DF 에 바로 추가하는 방법
    - 새로운 column 을 만들고 그안에 조건을 삽입.
    - d1['f']=d1['c']>5 

```python
d1 = pd.DataFrame( data=np.arange(16).reshape(4,4) , index=[10,20,30,40] , columns=list('abcd') , dtype=float)

d1
d1.loc[50]=[1,2,3,4]
d1['e']=[1,2,3,4,5]
d1

d1['f']=d1['c']>5 

d1['g']=np.where(d1['a']%3==0,1,10) # %3==0 이면 1, 0이 아니면 10 
d1
```

## column / row 삭제


#### del d1['a']

    - 자주 사용하지 않음.
    - 삭제되면 복구못함. 영구삭제 
    
#### drop ( 'row 또는 column 이름'  , axis=0 or 1 , inplace=True or False ) 

    - 자주 사용함 
    - axis 지정이 필수 
    - inplace = True 면 영구히 삭제, False 면 일시적으로 삭제 ( 주로 변수에 담을 때 사용 ) 

```python

d1
# columns 삭제 
# d1.drop(['d','e'], axis=1 , inplace=True) # 복수선택도 가능 

# row 삭제 
d1.drop([10,20] ,axis=0 , inplace=True) 
d1
```

## 실습2 
    
    - pandas_data3.txt 를 불러오시오.
    - 1.'gender' , 'final' columns을 삭제하시오
    - 2. 'lee7' , 'kim1' 를 삭제후 새로운 데이터 2개를 추가하시오
    
    - 3. 'rank' column  을 추가하시오, 이 때, year 가 1 또는 2 이면  L, year 3 또는 4 는 H 를 넣으시오 

```python
d1 = pd.read_table('pandas_data3.txt', index_col=0)

#1 

d1.drop(['gender','final'] , axis=1 , inplace=True )

# 2 

d1.drop(['lee7','kim1'] , axis=0 , inplace=True)

d1.loc['hyo']=[2,21,4,70]
d1.loc['bin']=[3,32,8,80]

d1


# 3  

d1['rank']=np.where(d1['year']>2,'H','L')
d1
```
