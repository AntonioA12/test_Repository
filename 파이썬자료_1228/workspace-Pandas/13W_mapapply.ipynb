## 연습 

    - 각 열별로 짝수들의 합이 3의 배수이면 G 
    - 3의 배수가 아니면 B 를 return 하시오.
    

```python
d1 = pd.read_table('pandas_data3.txt' , index_col=0)
d2 = d1.iloc[:3,1:]

def  G_B(x):
    x2 = x.loc[x%2==0]
    print (x2) 

    if np.sum(x2) %3 ==0 :
        return 'G'
    else:
        return 'B'
    
d2.apply(G_B)
```

    - 2. DataFrame 에서 10이상의 값은 'A' , 10미만 중 짝수는 'B' 10미만 중 홀수는 'C' 를 출력하라
    
    - 3. 각 학생 별 mid,final,proj 점수의 합이 140점 이상인 학생은 'P' 아니라면 'F'
    
    - 4. 각 열별 1)평균 2)표준편차 3)최대 4)최소 값을 구해서 DataFrame 에 추가하라 
    
    - 5. 각 학생 별 , final 이 final 평균이상 or mid가 mid 평균 이상이면 'P' 아니면 'F' 를 출력하라 

```python
d2 = d1.iloc[:10,1:]


#1 
def f1(x):
    m1 = np.mean(x)
    s1 = np.std(x)
    b1 = (((m1-s1)<x)&(x<(m1+s1)))
    return np.sum(x.loc[b1])

d2
d2.apply(f1)

#2 

def f2(x):
    if x >=10:
        return 'A'
    if x%2==0:
        return 'B'
    return 'C'
    
d2.applymap(f2)

```

```python
#3 

def f3(x):
    v1 = x.loc['mid']+x.loc['final']+x.loc['proj']
    
    if v1>= 140:
        return 'P'
    return 'F'

d2['P_of_F'] = d2.apply(f3, axis=1)
```
