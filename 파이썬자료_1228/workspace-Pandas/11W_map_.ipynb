## map

### 함수를 사용하지 않고 Series 코드 작성

    - 20보다 크도 50보다 작은 index의 평균을 구하라 

```python
s1 = pd.Series(np.random.randint(100,size=10))
s1

v1 = 20
v2 = 50
s2 = s1.loc[(s1>v1)&(s1<v2)]

s2.mean()
```

### 함수를 사용하고 Series 코드 작성

```python
def sum_btw_v1_v2(s1,v1,v2):
    
    s2 = s1.loc[(s1>v1)&(s1<v2)]
    
    return s2.mean()

sum_btw_v1_v2(s1,10,70)
```

### map 을 사용해보기 

```python
def f1(x):
    return (x+3)**2

def f2(x):
    if x>4:
        return 'L'
    elif x>2:
        return 'M'
    else :
        return 'S'

s1 = pd.Series(np.random.randint(10,size=5))
s1

s1.map(f1) # 모든 element 에 f1 함수가 적용된다
s1.map(f2)

```

## map 과 lambda 

#### lambda 
    
    - 한줄 짜리의 간단한 함수는 lambda 로 정의 및 활용 가능 ( 사칙연산 같은 ) 

```python
def f1(x):
    return (x+3)**2

s1.map(lambda x: (x+3)**2)
s1.map(f1)
(s1+3)**2 # < - 간단하긴 하나 제한적인 방법
```

### map 과 딕셔너리

    - 딕셔너리 key 에 조건을 부여해 , key 에 해당되는 값을 원하는 value로 변경할 수 있다.
    - 단, key 에 해당되지 않는 value 가 있다면, NaN 출력됨 ( 주의 )
    ( replace 함수는 없는 key 값은 원래 값을 유지 함 / NaN 이 없음 ) 

```python
s1 = pd.Series(np.random.randint(3 , size=6))
s1

s1.map({0:'a',1:'b',3:'c'})
s1.replace({0:'a',1:'b',3:'c'})
```

```python
def att_pass_or_faill(x):
    if x > 5 :
        return 'P'
    else:
        return 'F'
    
def proj_norm(x):
    return x*25/100

d1 = pd.read_table('pandas_data3.txt' , index_col=0)
d1 = d1.head(10)


d1['att_PF1'] = d1['att'].map(att_pass_or_faill)
d1['att_PF2'] = np.where(d1['att']>5,'P','F')
d1

d1['proj_norm1']=d1['proj'].map(proj_norm)
d1['proj_norm2']=d1['proj']*25/100
d1['proj_norm3']=d1['proj'].map(lambda x:x*25/100)
d1
d1['gender']= d1['gender'].replace({'F':1,'M':0})
d1
```
