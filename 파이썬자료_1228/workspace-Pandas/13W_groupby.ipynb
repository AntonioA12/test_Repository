## groupby 란?

    - DataFrame 의 columns , index 에 있는 항목끼리 분류 하는것이다.
    - EX) 남자 , 여자 끼리 묶음 / 1 학년 , 2학년 끼리의 묶음
    
    - 같은 element 들의 columns , index 들의 묶음

```python
d1=pd.read_table('pandas_data3.txt', index_col=0)
d1 = d1.head(10)
d1

d1.groupby('gender')
# F 끼리, M 끼리  / 출력타입은 turple() 
list(d1.groupby('gender'))

# Key = year , group = element
for key , group in d1.groupby('year'):
    print('Key=', key)
    group
```

## groupby  복수 그룹 / groupby([' gp1 ' , 'gr2']) 

```python

for key , gp in d1.groupby(['gender','year']):
    print( '#key ='  , key)
    gp
    
```

## get_group : groupby 에서 특정 그룹만 가져오기 

    - 남/여 중에서 여자만 불러오기
    
    - ( 남/여 ) , ( 1,2,3,4 학년 ) 중에서 ( 남자 ) ( 2학년) 만 가져오기

```python
d1.groupby(['gender']).get_group('F') 


d1.groupby(['gender','year']).get_group(('M',2))
```

## groupby  단수 그룹지정 후 , 그룹의 특정 columns 가져오기

    - 남자/여자  그룹에서 'mid' 점수를 가져오기

```python
for key , gp in d1.groupby('gender'):
    print(key)
    gp['mid']
    
# 둘이 똑같다   

for key , col in d1.groupby('gender')['mid']:
    print(key)
    col
```

## 복수 group 에서 특정한 값 가져오기 

    # (여자이고 1학년)인 mid 점수
    # (남자이고 2학년)인 mid 점수

```python

for key , gp in d1.groupby(['gender','year'])['mid']:
    print(key)
    gp
```

## groupby 와 apply ( 번외 : to_frame() ) 

    - groupby 된 변수에서 , apply 를 사용해 함수를 적용하는 방법이다.
    
    - 남/여 그룹의 mid 점수를 추출하고. np.sum() 사용하라
    
    - 번외 to_frame()
    - Series 를 Dataframe 으로 형변환

```python
for key , gp in d1.groupby('gender')['mid']:
    print(key)
    gp

d1.groupby('gender')['mid'].apply(np.sum)
d1.groupby('gender')['mid'].apply(np.sum).to_frame()

```

## grouby 와 apply ( 멀티 인덱스 ) / [  , ] 

    - ( 남자/여자) ( 1, 2, 3, 4, 학년 ) 그룹의 mid 점수의 합을 변수 s1 으로 저장.
    
    -  s1 과 loc 을 사용하여 
    -  ( 남자/여자) ( 1, 2, 3, 4, 학년 ) 그룹의 ( 특정 성별) , ( 특정 학년) 의 mid점수의 합을 가져온다

```python

for key , gp in d1.groupby(['gender','year'])['mid']:
    print( key )
    gp
    
s1 = d1.groupby(['gender','year'])['mid'].apply(np.sum)    
s1
s1.loc['F',1]
s1.loc['M',2]
s2 = d1.groupby(['gender','year'])['mid'].apply(np.sum).to_frame()
s2.loc['F',4]
s2.loc['M',3]
```

## grouby 와 apply ( 멀티 콜론 columns )  / [ [   ,   ] ]

    - ( 남자/ 여자 ) 그룹에서 [ 'mid' , 'final' ] 점수를 가져오기

```python
for key , gp in d1.groupby(['gender']):
    print ( key)
    gp[['mid','final']]

d1.groupby('gender')[['mid','final']].apply(np.sum)
```

## groupby : unstack  ( groupby and then groupby ) 

    - groupby() 된 DataFrame 을 key 별로 재 그룹 시키는것
    
    - ( 남자/여자 ) ( 1 ,2 ,3 ,4 학년) 그룹의 ['proj'] 의 평균을 나눈다.
    -  이후 , unstack() 사용하여 dataframe 으로 깔끔하게 표현

```python
d1=pd.read_table('pandas_data3.txt', index_col=0)

s1 = d1.groupby(['gender','year'])['proj'].apply(np.mean)
s1
s1.unstack()
```

## set_index & reset_index 

### set_index('인덱스 지정할 columns')
    
    - pd.read() 를 사용해서 index_col 없이 불러올때
    - set_index을 통해 원하는 colunms 을 index 로 변경 할 수 있다
    
    - set.index() 시, 기존 index 는 삭제 된다. 
    
### reset_index() 

    - set_index() 된 데이터를 초기화 해줌

```python
d1=pd.read_table('pandas_data3.txt')

d2 =d1.set_index('name')
d2.set_index('year')

d2.reset_index()
```

## 실습 

    - 1. gender 그룹 별로 mid 의 최대값은?
    
    - 2. year 그룹 별로  중간고사 점수가 28 이상인 학생 수 는?
    
    - 3. gender , year 그룹 별로  mid , final 의 평균은 ? 
    
    - 4. year 그룹 별로 final 평균 점수 이상의 점수를 받은 학생 수는? 
    
   

```python
#1 
d1=pd.read_table('pandas_data3.txt')

d1.groupby('gender')['mid'].apply(np.max)
```

```python
#2 

d1.groupby('year')['mid'].apply( lambda x: np.sum(x > 28 ) )
```

```python
# 3 

d1.groupby(['gender','year'])[['mid','final']].apply(np.mean).round(2)
```

```python
# 4 

d1.groupby('year')['final'].apply( lambda x: np.sum(x > d1['final'].mean() ) )
```
