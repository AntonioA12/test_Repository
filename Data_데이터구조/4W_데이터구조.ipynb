## 연결리스트

    순서가 있는 데이터를 나열한 자료구조 
    
## 노드 

    연결리스트를 구성하는 하나의 원소
    
    data , pointer ( 자료 , 방향 ) 2가지로 구성
    포인터 = 다음 노드의 방향 
    
    자기참조
        자신과 같은 형의 인스턴스를 참조하는 필드가 있는 구조 ( 속성 )  
        
    타입힌트
        ex ) a : int , b : float , C : str  
        함수의 매개변수와 리턴 자료형을 명시한다.
        강제성은 없다. ( 타입형을 틀려도, 문법만 맞는다면 실행은 된다 )  
        
        ) -> 리턴형 
        b : int ) -> float : 
        int 를 리턴해도 되고, float 를 리턴해도 된다 
    
## 타입종류    
 
     Any 모든타입
     int 정수
     float 실수
     str 문자열 
     boolean T/F
     
     List
     List[int] 정수를 원소로 가진 리스트 
     List[str] 문자열을 원소로 가진 리스트 
     
     Tuple
     Tuple[int,str, int ]  정수 문자열 정수를 원소로 가지는 튜플 
     
     Dict
     Dict[str,int] key = 문자열, value = 정수형 
     

````python
from __future__ import annotations
from typing import Any,List,Tuple,Dict  # 타입힌트 라이브러리 

class Node:
    """연결 리스트용 노드 클래스"""
    
    def __init__ ( self , data: Any = None , next: Node = None )-> None: # 마지막 노드의 next(포인터)는 리턴이 없다. = None 그자체  아무것도 없다는 뜻  
        """초기화"""
        self.data = data
        self.next = next 
````

````python
n0 = Node('A')
n1 = Node('B')
n2 = Node('C')

n0.next = n1
n1.next = n2

print(n0.data)
print(n0.next.data)
print(n0.next.next.data)
````

````python
n2 = Node('C')
n1 = Node('B', n2) # ( data , pointer  )
n0 = Node('A' , n1 ) # ( data , pointer)

print(n0.data)
print(n0.next.data)
print(n0.next.next.data)
````
