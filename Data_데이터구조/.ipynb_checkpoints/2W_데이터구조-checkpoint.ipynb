## 데이터구조

    자료구조   
    알고리즘   
    알고리즘 성능평가방법 
    

## 추상데이터 -> 자료구조 - > 알고리즘
## 자료구조 

    자료를 정리하고 조직화 
    ( 편리하고 효율적으로 사용하기 위해 , 효율적 규칙 ) 
    
    스택, 큐 , 리스트 , 딕셔너리 , 탐색구조 , 그래프(조직도)  , 트리 ( tree = all )  
    
    데이터를 정돈하여 저장한 구성체 
    
    데이터 + 데이터 관련된 연산들을 함께 고려해야한다 ( 수학 ) 
    
### 추상데이터 ( 기획서 ) 
    
    구체적이지 않음.
    추상 = 연산을 구체적으로 어떻게 구현하여하 한다는 세부 명세를 포함하지 않은 데이터 
    
    데이터 + 데이터 추상적인 연산들로 구성
    
    자료구조는 추상데이터타입을 구체적(실제 프로그램)으로 구현한것
    
    [ 추상데이터 = 기획서, 설계도 
    자료구조 = 결과물  ]
    
## 알고리즘 

    프로그램 = 자료구조 + 알고리즘 
    알고리즘 = 문제를 해결하기위한 단계적인 절차 
    
    절차 1 - 절차 2 - 절차 3 - 절차 4 - 절차 5 
    
    [(데이터 + 연산 ) = 절차 ( 자료구조 )]
    
    효율적 알고리즘 - 간단하고 빠르면서 목적수행이 가능한 
    알고리즘 성능 
    1. 수행시간을 나타내는 시간복잡도 + 사용되는 메모리 양  ( 대부분 시간을 우선시함 ) 
    2. 공간의 크기를 나타내는 공간복잡도 
    
    시간 > 메모리양 > 공간 
    
    2가지에 기반하여 분석 
    
###  시간복잡도 

     알고리즘이 실행될 때, 기본적인 연산횟수를 입력크기의 함수로 나타냄
     ( 기본연산 = 데이터간의 크기 비교, 읽기 및 갱신 , 숫자 계산 등 ) 
    
    f(N) =  0 + 0 + 0 
    N = 연산 횟수 
    
    1. 최악경우 분석
        가장 빠른 수행시간을 분석 
    2. 평균경우 분석
        입력의 확률 분포를 가정하여 분석, 균등분포로 가정한다 
    3. 최선경우 분석
        어떤 입력이 주어지더라도 알고리즘이 수행시간이 얼마 이상은 넘지 않는다는 '상한' 의 의미 
    
        운으로 결정된다 ? 
        60을 만들기 위해 주사위를 몇번 던져야 할까 ?
        최선경우 - 6이 10번 
        최악경우 - 1이 60번
        평균경우 - 60 / 3.5 = 17
        
        일반적으로 알고리즘의 수행시간은 최악경우 분석으로 표현 
        항상 최악의 경우를 가정하고 대비해서 분석을한다. ( 성능을 평가하기 위해 ) 
        [ 최악의 경우보다 더 낮은 경우는 없으니까 ] 
        
        어떤 알고리즘이 더 효율적이고 덜 효율적인가 ? 
    

### 접근 표기법 3가지 


    1. O - 표기법 
    
        모든 N > N0 에 대해서 f(N) < cg(N) 이 성립하는 양의 상수 c와 N0이 존재하면,
        f(N) = O(g(N)) 이다. 
        
        O(N) = O(logN) 이라면 O(logN) 표기법을 사용해야한다. 
        
        
       N0 와 같거나 큰 모든 N 에 대해서 f(N) 이 cg(N)보다 크지않다는 것
       
       f(N) = O(g(N)) 은 N0 보다 큰 모든 N 에 대해서 f(N)이 양의 상수를 곱한 g(N)에 미치지 못한다는 뜻
       
       g(N)을 f(N)의 상한(Upper Bound)이라고 한다. [ g(N) = 상한 =  비고(Big O?)   =  g(N) 위로는 못 올라간다. ] 
       cg(N)은 상한이다. ( 최대값 ) 
       
       ex ) 
       
           f(N) = 2N**2 + 3N + 5 
           
           g(N) = N**2 으로 정하면 
           3보다 큰 모든 N에 대해 2N**2 + 3N + 5 < 4N**2 
           N0= 3, C= 4 일떄, 
           2N**2 + 3N + 5 < 4N**2 = f(N) = O(N**2) 이 되고, 
           
           O(N**2) 으로 표기할 수 있다 ???? 
           이때, 가장 차수가 낮은 함수를 선택하는것이 바람직하다 
           ( O(N**2) , O(N**3), O(N**4) 가 있다면, O(N**2) 을 선택하자 ) 
       
       ex 2 )
       
           2N**2 + 3N + 5  식이 있다면,
           최고차항 2N**2을 표기하고 
           나머지는 무시한다 
           
           O(N**2) 으로 표기하고, 평가한다. 

    2. 오메가 표기법 
    
        O =  더이상 위로 올라가지않음
        오메가 = 더이상 내려가지 않음 
        
        f(N) > cg(N) 이 성립한다면 f(N) = 오메가(g(N))으로 표기할 수 있다. 
        오메가 = 하한 
        
        가장 낮은 높은 차수의 함수를 선택하는것이 바람직. 
        

    3. 세타 표기법 
    
        보다 더 정확하게 표현 ( 동일한 증가속도로 표현 ) 
        O 표기법 이고, 동시에  오메가 표기법이 적용된다면 
        f(N) = 세타(g(N)) 이다. 
        
        상한, 하한 모두 표기 
        일정한 증가율이고 기울기가 일정하다 = 세타 
        
        
## 표기법 요약 

    1. O(1) 상수시간 
        입력값이 아무리 커도 실행시간은 일정 ( 10 입력하는 1억번 입력하든 시간이 일정 ) 
        이론상 최상의 알고리즘 
        
    2. O(logN) 로그시간 
        실행시간은 입력값에 영향을 받지만, 웬만한 큰 N에 대해서도 매우 빠르다.
        이진검색 알고리즘 수행시간 
        ( 입력수 up = 수행시간 up , but 그래도 빠른편 ) 
        
    3. O(N) 선형시간
        입력값에 비례해서 수행시간이 증가한다. 
        ( 10개 검색 = 10분 , 100개 검색 = 100분 ) 
        
    4. O(N * logN) 로그선형시간 
        대부분 효율좋은 정렬 알고리즘의 수행시간
        비교 기반정렬 알고리즘은 아무리 좋은 알고리즘도 O(N logN) 보다 빠를 수 없다. 
        
    5. O(N**2) 제곱시간
        버블정렬 같은 비효율적인 알고리즘 수행시간
    
    6. O(N**3) 세제곱시간
          여기까진 쓸만하긴 함 
          
    7. O(2**n) 지수시간 
        비효율적
    8. O(N!) 계승시간 
       매우 비효율적
