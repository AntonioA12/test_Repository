## 스택

    후입선출
    가장 마지막에 넣은 데이터가 
    가장 먼저 추출된다.

#### 탑 TOP

    가장 최근에 들어온 데이터 ( 맨 위 )

#### 바텀 Bottom

    가장 마지막에 들어온 데이터 ( 맨 아래 ) 

#### 푸시 Push

    데이터 입력 

#### 팝 Pop

    데이터를 꺼냄으로써 삭제

````python
from __future__ import annotations
from typing import Any, List, Tuple, Dict
````

````python
class Stack:
    
    def __init__(self) -> None:
        """ 스택 생성 """
        self.stack = [] # 오른쪽 끝이 top 이다 , [-1] = top 
        
    def is_empty() -> bool:
        """ 스택이 비어있는지 확인 """
        return len(self.stack) == 0 # 길이가 0 이면 리턴 
    
    def pop(self)-> Any:
        """ 스택에서 탑에 있는 데이터를 꺼내기 """
        if not self.is_empty(): # 스택이 비어있지 않으면 
            return self.stack.pop() # 가장 최근 데이터 꺼내기 ( + 동시에 삭제 ) 
        else:
            raise IndexError(' 스택이 비어있습니다. ')
            
    def push(self, data:Any):
        """ 스택에 데이터 삽입 """
        self.stack.append(data)
    
    def peek(self) -> Any:
        """ 스택에서 탑에 있는 데이터를 리턴 """
        if not self.is_empty():
            return self.stack[-1] # top의 데이터를 리턴 
        else:
            raise IndexError(' 스택이 비어있습니다. ')
    
    def print(self)-> None:
        """ 스택의 모든 데이터를 출력 """
        print(f'{self.stack} : Top, (맨 오른쪽이 top입니다)')
    
    def clear(self)-> None:
        """ 스택의 모든 데이터를 삭제"""
        while not self.is_empty(): # 스택이 비어있지 않으면 
            self.pop() # 계속해서 pop 함수 실행 
    
    def __len__(self)-> int:
        """스택의 크기(길이) 리턴"""
        return len(self.stack)
            
````

````python
stack= Stack()
stack.push(1)
stack.push(2)
stack.push(3)

print(stack.pop()) 
stack.push(4)
print(stack.peek())
stack.push(5)
print(stack.pop())

print('*'*10)
stack.print()
print(len(stack))
````

## 푸시 팝 픽 len = O(1) 이다 
## clear 최악의 경우 O(N) 

````python
stack_1 = list()
stack_1.append(1) # push
stack_1.append(2)
stack_1.append(3)

print(stack_1.pop()) # 3
stack_1.append(4)
print(stack_1[-1]) # peek 부분
stack_1.append(5)
print(stack_1.pop())

print('aa'*10)

print(stack_1 , ": Top")
print(len(stack_1))

stack_1.clear()
print(stack_1,':Top')
````

## 스택을 이용하여 괄호 짝 맞추기
    
    알고리즘 
       왼쪽 ( 는 push , 오른쪽 ) 은 pop
       pop된 왼쪽 ( 와 바로 읽었던 오른쪽 ) 가 다른 종류이면 False 리턴 ( ex. (] )
       같은 종류라면 True ( ex. [] ) 
       
    모든 괄호를 읽은 뒤 스택이 비어있으면 True 리턴 
    
    
### 예제 

    () - True 
    )( - False 
    ()[] - True 
    ( [] ) - True
    ( [ ) ) - False 
    
    (( [] {} )) 
    True 인 외곽은 저장 
    닫힌 [] {} 을 처리 ( 삭제 함 으로써 )
