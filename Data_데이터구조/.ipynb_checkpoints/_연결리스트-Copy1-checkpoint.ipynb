## 연결리스트 구현


### ptr = self.head 가 없는 함수

    __init__
    __len__
    remove_first ( 사실 없다기보단 다른 지정을 함 ) 
    __contains__
    
### while ptr.next is not None: 으로 반복이동해야하는 함수

    add_last()
    print() # while ptr is not None : ( 왜? while 문의 있는 ptr.함수를 따로 또 지정해야함 ) 
    search() # while ptr is not None: ( 왜? while 문의 있는 ptr.함수를 따로 또 지정해야함 ) 
    remove_last() # ( 얜 pre 까지 지정 ) 
    remove()
    

````python
from __future__ import annotations
from typing import Any, List, Tuple, Dict

class Node:
    """연결 리스트용 노드 클래스"""

    def __init__(self, data: Any = None, next: Node = None):
        """초기화"""
        self.data = data  # 데이터
        self.next = next  # 뒤쪽 포인터


class LinkedList:
    """연결 리스트 클래스"""

    def __init__(self) -> None:
        """초기화"""
        self.no = 0          # 노드의 개수
        self.head = None     # 머리 노드

    def __len__(self) -> int:
        """연결 리스트의 노드 개수를 반환"""
        
        return self.no 
        


    def add_first(self, data: Any) -> None:
        """맨 앞에 노드를 삽입"""
        
        ptr = self.head
        self.head = Node( data , ptr)
        self.no += 1 

    def add_last(self, data: Any):
        """맨 끝에 노드를 삽입"""
        if self.head is None:
            self.add_first(data)
        else:
            ptr = self.head
            
            while ptr.next is not None:
                ptr= ptr.next 
            ptr.next = Node( data , None)
            self.no += 1 
        

    def print(self) -> None:
        """모든 노드를 출력"""
        
        ptr = self.head 
        
        while ptr is not None:
            print( ptr.data , end='')
            if ptr.next is not None:
                print('->' , end='')              
            else:
                print()
            ptr = ptr.next 
        
    
    def search(self, data: Any) -> int:
        """data와 값이 같은 노드를 검색"""
        
        idx = 0 
        ptr = self.head 
        
        while ptr is not None:
            if ptr.data == data:
                return idx 
            idx += 1 
            ptr = ptr.next 
        return -1 
        
    def __contains__(self, data: Any) -> bool:
        """연결 리스트에 data가 포함되어 있는가?"""
        return self.search(data) >= 0 

    def remove_first(self) -> None:
        """머리 노드를 삭제"""
        
        if self.head is not None:
            self.head = self.head.next 
            self.no -= 1 
        

    def remove_last(self):
        """꼬리 노드 삭제"""
        # 노드가 있다면 
                

        if self.head is not None:
            
            # 노드가 1개일 경우를 대비해 remover_first() 호출
            if self.head.next is None:
                self.remove_first()
            else:
                
                ptr = self.head # 꼬리 노드 
                pre = self.head # 꼬리 앞 노드 
                
                while ptr.next is not None: # ptr 뒤에 노드가 있으면 
                    
                    pre = ptr
                    ptr = ptr.next  # 노드 한칸 이동 
                    
                pre.next = None
                self.no -= 1 
        
        

    def remove(self, p: Node) -> None:
        """노드 p를 삭제"""
  
        if self.head is not None: 
            if p is self.head:
                self.remove_first()
            else:
                ptr = self.head 
                
                while ptr.next is not p:
                    ptr = ptr.next 
                    if ptr is None:
                        return
                ptr.next = p.next
                self.no -= 1
                
        
    def clear(self) -> None:
        """전체 노드를 삭제"""
        
        while self.head is not None:
            self.remove_first()
        self.no = 0 

    def __iter__(self) -> LinkedListIterator:
        """이터레이터(반복자)를 반환"""
        return LinkedListIterator(self.head)
    
    def count(self, data: Any) -> int:
        """연결 리스트에 data가 포함되어 있는 노드의 개수"""
        """ A A C D C C 연결리스트에서 A 의 갯수 B 의 갯수 C 의 갯수 D 의 갯수 출력하기 
        
        count = 0 
        ptr = self.head 
        
        while ptr is not None:
            if ptr.data == data:
                count +=1
            ptr = ptr.next
        return count
        


class LinkedListIterator:
    """클래스 LinkedList의 이터레이터(반복자)용 클래스"""

    def __init__(self, head: Node) -> None:
        self.current = head

    def __iter__(self) -> LinkedListIterator: # iter() 함수 호출시 자동 실행
        return self

    def __next__(self) -> Any: # next() 함수 호출시 자동 실행
        if self.current is None:
            raise StopIteration
        else:
            data = self.current.data
            self.current = self.current.next
            return data
        
## 아래는 수정하지 마시오.


my_list = LinkedList()
my_list.add_last('A')
my_list.add_last('B')
my_list.add_last('C')
my_list.add_last('D')

for i in my_list:
    print(i)

my_list.print()
my_list.remove_last() # 마지막 노드 제거
my_list.print()
my_list.remove(my_list.head.next) # 머리노드 다음 노드 제거
my_list.print()

print('*'*20)

my_list = LinkedList()
my_list.add_last('A')
my_list.add_last('A')
my_list.add_last('C')
my_list.add_last('D')
my_list.add_last('C')
my_list.add_last('C')

print(my_list.count('A'))
print(my_list.count('B'))
print(my_list.count('C'))
print(my_list.count('D'))
````

# 문제2

파이썬의 range 클래스를 모방하여 for문에서 사용가능한 my_range 클래스를 작성하시오.

start 에서 시작하여 stop 이전에 끝나게 되도록 구현하시오.

***결과***
```
1
2
3
-------
0
1
2
3
4
```

````python
class my_range:
    def __init__(self, start: int, stop: int):
        self.start = start
        self.current = start
        self.stop = stop
        
    def __iter__(self) -> my_range: # __next__() 메소드를 가진 객체(반복자)를 리턴해야 한다.
        return self
        pass
    
    def __next__(self) -> int:
        if self.current == self.stop:
            raise StopIteration
        else:
            data = self.current
            self.current += 1
            return data
        pass
        
# 아래는 수정하지 마시오.
for i in my_range(1, 4):
    print(i)

print('-------')    

for i in my_range(0, 5):
    print(i)
````
