## 트리 

#### 트리 란?
    트리를 이용하면 [계층적 데이터] 를 사용할 수 있고, [계층적 데이터] 효율적으로 다룰 수 있다.
    ex) 파일 시스템, 데이터 베이스 , 웹 사이트, 생물의 진화 , 조직도, 게임 스킬트리 등 모든 분야에서 사용
    
    상위 트리 - 하위 트리
    B - 트리 
    
#### 노드 와 엣지

    노드 = 점 
    엣지 = 선
    
    엣지의 방향은 상위 -> 하위 트리 순서로 이동
    방향표는 생략하고 선만 이어져 있다. 
    



## 트리 용어 ( 중요 , 객관식 문제로 많이 나올 듯  )

    트리 인 경우 
    재귀적 정의
        1.노드, 엣지 하나도 없이 비어있어도 트리이다 ( null tree)
        2.엣지 없이 노드 하나만 있어도 트리이다. 
        3.트리에 자식노드를 추가한 것은 트리이다.
        
    비재귀적 정의 
        사이클(루프)가 없는 연결된 그래프
        
    트리가 아닌 경우
        1. 1노드 - 2부모 일 경우 
        2. 사이클(루프) : 노드 끼리 연결되어 무한루프되는 경우 
        3. 노드끼리 엣지가 분리 되어있는 경우 
        
    루트노드 
        트리에서 최상위(가장 위) 에 있는 노드 ( 시작점 ) 
        어떤 트리도 2개의 루트를 가질 수 없다.
        루트노드 는 전체 트리를 대표한다. 
        
        루트 노드 이름이 A 라면
        트리 A 라고도 불린다. 
        
    부모 노드 
        루트 노드를 제외한 모든 노드는 부모 노드를 가지고 있다. 
        상위 - 하위 노드 관계를 가진 노드 
        
    자식 노드
        루트 노드를 제외한 모든 노드는 자식노드를 가질 수 있다.
        ( 최 하위 노드는 제외, 자식노드 X ) 
        
    형제 노드 
        같은 부모노드를 가지는 노드 그룹을 형제노드  
    조상 노드
        특정 노드 가 위쪽으로 올라가면서 만나는 모든 노드들 ( 부모의 부모의 부모 노드 ) 
    자손 노드
        특정 노드 기준으로 하위에 있는 모든 노드들 ( 자식의 자식 의 자식 노드 ) 
    리프 노드 
        자식이 없는 최하위 노드 ( 무조건 맨 마지막 노드가 아닌 자식이 없는 노드 )  
    내부 노드
        자식이 있는 노드 ( 리프 노드와 반대 )
        비단말 노드 라고도 함 
    레벨 노드 
        트리의 층수, 얼마나 멀리, 얼마나 깊은지 확인하는 방도 
        [가장 최상층이 루트 레벨은 0이다.]
        계층이 내려갈 수록 레벨 +1 
    차수( degree ) 
        노드가 갖는 자식의 수 
    노드의 높이
        특정 노드에서 가장 멀리 있는 리프노드 까지의 거리
        리프노드와의 거리 
        
        바닥 기준 = 리프노드
        천장 기준 = 특정 노드
        
        자식이 없는 리프노드는 높이가 0 
    트리의 높이
        루드 노드의 높이 = 트리의 높이 = 동시에 레벨의 최대 값 
    서브 트리 
        전체 트리에서 특정 노드를 루트노드로 지정하고
        지정된 특정 노드( 제2의 루트노드 ) 가 가진 모든 노드들을 서브 트리로 지정 ( 트리 분할 ? ) 


## 이진 트리
    이진 트리 = 2개 이하의 자식노드를 가지는 트리 
    특징 : 왼쪽 자식 / 오른쪽 자식으로 자식 노드를 구분함 
    
    특정 노드 기준으로 
    왼쪽 서브 트리 
        왼쪽 자식은 왼쪽자식 루트 = 서브트리,
    오른쪽 서브 트리 
        오른쪽 자식 루트는 오른쪽 서브트리이다.
        
    포화 이진트리
        모든 레벨이 노드들로 꽉 차있는 이진트리 
    완전 이진트리 ( 거의 완전한 이진트리 , almost complete binary tree , 주로 완전 이진트리 사용할 것 )
        마지막 레벨을 제외한 각 레벨들이 노드들로 꽉 차있고, 
        
    배열을 이용한 이진트리 구현
        1. 인덱스 0 인 자리는 비워둔다
        2. 루트노드의 인덱스 번호는 1이다.
        
#### 트리 탐색
    
    너비 우선 탐색
        Queue(큐)를 사용하여 루트노드 로 시작하여 왼쪽 -> 오른쪽 방향으로 노드를 탐색하는 방법 
    깊이 우선 탐색 
        스택(stack + 재귀 ) 를 사용하고 세로검색, 수직검색 방향
