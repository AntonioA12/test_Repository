```python
from __future__ import annotations
from typing import Any, List, Tuple, Dict

class Node:
    """연결 리스트용 노드 클래스"""

    def __init__(self, data: Any = None, next: Node = None):
        """초기화"""
        self.data = data  # 데이터
        self.next = next  # 뒤쪽 포인터


class LinkedList:
    """연결 리스트 클래스"""

    def __init__(self) -> None:
        """초기화"""
        self.no = 0          # 노드의 개수
        self.head = None     # 머리 노드

    def __len__(self) -> int:
        """연결 리스트의 노드 개수를 반환"""
        
        return self.no 
        


    def add_first(self, data: Any) -> None:
        """맨 앞에 노드를 삽입"""
        
        ptr = self.head
        self.head = Node( data , ptr)
        self.no += 1 

    def add_last(self, data: Any):
        """맨 끝에 노드를 삽입"""
        if self.head is None:
            self.add_first(data)
        else:
            ptr = self.head
            
            while ptr.next is not None:
                ptr= ptr.next 
            ptr.next = Node( data , None)
            self.no += 1 
        

    def print(self) -> None:
        """모든 노드를 출력"""
        
        ptr = self.head 
        
        while ptr is not None:
            print( ptr.data , end='')
            if ptr.next is not None:
                print('->' , end='')              
            else:
                print()
            ptr = ptr.next 
        
    
    def search(self, data: Any) -> int:
        """data와 값이 같은 노드를 검색"""
        
        idx = 0 
        ptr = self.head 
        
        while ptr is not None:
            if ptr.data == data:
                return idx 
            idx += 1 
            ptr = ptr.next 
        return -1 
        
    def __contains__(self, data: Any) -> bool:
        """연결 리스트에 data가 포함되어 있는가?"""
        return self.search(data) >= 0 

    def remove_first(self) -> None:
        """머리 노드를 삭제"""
        
        if self.head is not None:
            self.head = self.head.next 
            self.no -= 1 
        

    def remove_last(self):
        """꼬리 노드 삭제"""
        # 노드가 있다면 
                

        if self.head is not None:
            
            # 노드가 1개일 경우를 대비해 remover_first() 호출
            if self.head.next is None:
                self.remove_first()
            else:
                
                ptr = self.head # 꼬리 노드 
                pre = self.head # 꼬리 앞 노드 
                
                while ptr.next is not None: # ptr 뒤에 노드가 있으면 
                    
                    pre = ptr
                    ptr = ptr.next  # 노드 한칸 이동 
                    
                pre.next = None
                self.no -= 1 
        
        

    def remove(self, p: Node) -> None:
        """노드 p를 삭제"""
  
        if self.head is not None: 
            if p is self.head:
                self.remove_first()
            else:
                ptr = self.head 
                
                while ptr.next is not p:
                    ptr = ptr.next 
                    if ptr is None:
                        return
                ptr.next = p.next
                self.no -= 1
                
        
    def clear(self) -> None:
        """전체 노드를 삭제"""
        
        while self.head is not None:
            self.remove_first()
        self.no = 0 

    def __iter__(self) -> LinkedListIterator:
        """이터레이터(반복자)를 반환"""
        return LinkedListIterator(self.head)
    
    def count(self, data: Any) -> int:
        """연결 리스트에 data가 포함되어 있는 노드의 개수"""
        """ A A C D C C 연결리스트에서 A 의 갯수 B 의 갯수 C 의 갯수 D 의 갯수 출력하기""" 
        
        count = 0 
        ptr = self.head 
        
        while ptr is not None:
            if ptr.data == data:
                count +=1
            ptr = ptr.next
        return count
        


class LinkedListIterator:
    """클래스 LinkedList의 이터레이터(반복자)용 클래스"""

    def __init__(self, head: Node) -> None:
        self.current = head

    def __iter__(self) -> LinkedListIterator: # iter() 함수 호출시 자동 실행
         return self

    def __next__(self) -> Any: # next() 함수 호출시 자동 실행
        if self.current is None:
             raise StopIteration
        else:
            data = self.current.data
            self.current = self.current.next
        return data
        
## 아래는 수정하지 마시오.


my_list = LinkedList()
my_list.add_last('A')
my_list.add_last('B')
my_list.add_last('C')
my_list.add_last('D')

for i in my_list:
    print(i)

my_list.print()
my_list.remove_last() # 마지막 노드 제거
my_list.print()
my_list.remove(my_list.head.next) # 머리노드 다음 노드 제거
my_list.print()

print('*'*20)

my_list = LinkedList()
my_list.add_last('A')
my_list.add_last('A')
my_list.add_last('C')
my_list.add_last('D')
my_list.add_last('C')
my_list.add_last('C')

print(my_list.count('A'))
print(my_list.count('B'))
print(my_list.count('C'))
print(my_list.count('D'))
```

```python
from __future__ import annotations
from typing import Any, List, Tuple, Dict

class Node:
    """연결 리스트용 노드 클래스"""

    def __init__(self, data: Any = None, next: Node = None):
        """초기화"""
        self.data = data  # 데이터
        self.next = next  # 뒤쪽 포인터


class LinkedListStack:
    """연결리스트를 이용한 스택 클래스"""

    def __init__(self) -> None:
        """초기화"""
        self.no = 0          # 노드의 개수
        self.top = None     # self.head = self.top 

    def __len__(self) -> int:
        """스택의 크기 반환"""
        return self.no
    
    def is_empty(self) -> bool:
        """스택이 비어있는지 확인"""
        return self.top is None 


    def push(self, data: Any) -> None:
        """스택에 데이터를 푸시"""

        ptr = self.top 
        self.top = Node (data , ptr )
        self.no += 1 

        
    def pop(self) -> Any: # top을 삭제하고 리턴  
        """스택에서 탑에 있는 데이터를 꺼냄(삭제)"""

        if not is_empty():
            data = self.top.data  # 리턴 ( 출력 부분 )
            self.top = self.top.next  # 출력 후 삭제 
            self.no -= 1 
        else:
            raise IndexError('스택비어있습니다')
        
    def peek(self) -> Any: # 삭제 안하고 top 을 리턴 
        """스택에서 탑에 있는 데이터를 반환"""
        
        if not self.is_empty(): 
            return self.top.data # 출력만  
        else:
            raise IndexError( '스택이 비어있습니다 ')
    
    def print(self) -> None:
        """스택의 모든 데이터를 탑에서 부터 출력"""
        ptr = self.top
        print('Top=', end='')
        
        while ptr is not None:
            print(ptr.data, end='')
            if ptr.next is not None:
                print('->', end='')
            else:
                print()
            ptr =ptr.next 
            
    def clear(self) -> None:
        """전체 노드를 삭제"""
        while not self.is_empty():  # 전체가 비어 있게 될 때까지
            self.pop() # 팝 연산
        self.no = 0
        
stack = LinkedListStack()
stack.push(1)
stack.push(2)
stack.push(3)
print(stack.pop()) # 3
stack.push(4)
print(stack.peek()) # 4
stack.push(5)
print(stack.pop()) # 5
print("----")
stack.print()
print(len(stack))
print("----")
    
stack.clear()
stack.print()

```
