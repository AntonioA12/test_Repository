## 큐 Queue 

    선입선출( Fisrt In - First Out ) 
        가장 마지막에 넣은 값이 가장 먼저 나오는 데이터 입출력 방식 
        
        -> D - >  C -> B -> A 
        
        줄에서 데이터가 춥력되는곳이 앞( Front )이다 
        줄에서 데이터가 입력되는곳이 뒤( Rear ) 이다. 
        
    구현방법
    1.연결리스트
    2.파이썬 데크 
    3. 파이썬 데크를 큐처럼
    
    활용도는 스택 < 큐 
    활용 예시 = 일방통행 , 줄서서 업무대기 , 
    윈도우 메세지 처리 , 가장 빠른 길 찾기 ( 큐 응용 ) 
        
### Inqueue 인큐( 입력)  와  Dequeue 디큐 ( 출력 ) 
    
    
     디큐 ( 스택에서 팝 ) 
     데이터를 꺼내는 작업 ( 꺼내면서 삭제 ) 
     ( 삭제하고 리턴 하는것은 pop 과 동일 )  
    

## 연결리스트와 큐Q

    연결리스트 와 큐의 방향 이 반대 
    
                # 연결리스트 
    <<(앞)       [ A->B->C->D->E ]        <<(뒤) 
                A.next = B 
              

````python
from __future__ import annotations
from typing import Any, List, Tuple, Dict
````

````python
class Node:
    """연결 리스트용 노드 클래스"""

    def __init__(self, data: Any = None, next: Node = None):
        """초기화"""
        self.data = data  # 데이터
        self.next = next  # 뒤쪽 포인터
        
class LinkedListQueue:
    
    
    def __init__(self) -> None:
        """비어있는 큐 생성자 """ # 빈상자 
        
        self.no = 0
        self.front = None
        self.rear = None
        
    def __len__(self) -> int:
        """ 큐의 크기 리턴"""
        return self.no 
    
    def is_empty(self) -> bool:
        """큐가 비어있는지 체크"""
        return self.front is None # 비어있다면 T 
    
    def enqueue(self, data:Any)-> None:
        """큐에 rear에 데이터 삽입"""
        " O(1) "
        
        node = Node(data) # New 데이터
        
        if self.is_empty():
            self.front = None
            self.rear = None
            
        else: # 선입선출이기 때문에 rear(뒤) 위치에서 데이터가 입력된다. 
            self.rear.next = node  # 노드 = 새로운 데이터 입력(Linking)
            self.rear = self.rear.next # rear 업데이트 #  
        self.no +=1 
        
    def dequeue(self) -> Any:
        """ 큐에서 데이터를 출력(동시에 삭제)"""
        " O(1) "
        if not self.is_empty(): # 큐가 비어있지 않으면 
            data = self.front.data  # 프런트 데이터를 저장하고
            self.front = self.front.next # A (삭제).next -> B # 프런트 노드를 삭제하고 데이터를 리턴 
            self.no -= 1
            return data # 삭제 후 , 데이터 리턴 
        
        else: # 비어있으면 출력이 불가능 
            raise IndexError('큐가 비어있습니다. ')
    
    def peek(self) -> Any:
        """큐에서 프런트에 있는 데이터를 반환, 삭제하면 안됨 """
        " O(1) "
        
        if not self.is_empty(): # 데이터가 있으면 
            return self.front.data # 프론트 데이터 리턴 
        else:
            raise IndexError('큐가 비어있습니다')
            
    def print(self) -> None:
        """ 큐의 모든 데이터를 프런트->rear 순서로 출력"""

        ptr = self.front # end='' 줄바꿈 안함 
        print("Front : = " ,end='') # 프론트 1개 출력 
        
        while ptr is not None: # 
            print( ptr.data, end='') # 데이터 출력
            if ptr.next is not None: # 계속해서 데이터가 있으면 
                print( "->" , end='' ) 
            ptr = ptr.next # 반복 
        print('  = Rear' ) # while 종료 = 맨 끝 
        
    def clear(self)-> None:
        """전체노드를 삭제 """
        while not self.is_empty(): # 데이터가 있으면 
            self.dequeue() # 계속해서 삭제 
        self.no = 0
        
        
````
