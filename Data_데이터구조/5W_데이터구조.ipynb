### 삭제 remove(): 임의의 노드 삭제 # p노드가 머리 노드라면 O(1), 아니라면 이동할수록 N-1 되고, 항상 최악의 경우를 생각하기 때문에 O(N) 

    특정 노드를 삭제하는 함수 
    
    특정 노드가 머리노드라면 remove_fisrt():
    머리노드가 아니라면 특정노드 앞을 찾아
    앞 노드 .next 를 통해 삭제 
    
    ptr = 앞 노드 
    while 문 = ptr의 위치를 찾기 위함
        ptr의 위치를  찾으면 종료 
        
```python
        def remove(self, p:Node)-> None:
        """특정 노드 = p 노드  삭제"""
        
        if self.head is not None: 노드가 비어있지 않으면, 
            if p is self.head:  p 노드가 머리노드라면 
                self.remove_first() 머리노드삭제 함수 실행 
            else:
                ptr = self.head p노드 앞 노드를 지정 
                
                
                while ptr.next is not p: p노드가 삭제되었다면 
                    ptr = ptr.next p 노드 앞을 한칸 이동하고 ( 삭제한 부분을 연결시키는 작업 ) 
                    if ptr is None: p 노드 가 존재하지 않으면 return 함으로써 while 문 종료 
                        return 
                    
                    ptr.next = p.next p 노드 = 삭제 
                    self.no -= 1 노드수 -1 

```

## for문은 어떻게 동작하는가? 

    1.__iter__() 메소드를 호출하여 반복 ( 1번호출 ) 
    2. 반복자 객채의 __next__() 메소드를 호출하여 다음 원소를 얻는다. ( 원소의 갯수만큼 호출=반복 )
    3. 예외가 발생할 때까지 2를 반복한다. 
    
    
#### iterable 반복가능객체 

    __iter__() 메소드를 가지고 있는 객체 
    
    for item in iterable: 
    for 문에 사용가능
    반복가능 객체 , 리스트 또는 튜플로 구성 
    
#### iterator 반복자 

    __next__() 메소드를 가지고 있는 객체 
    다음 값을 가져올수 없다면 예외를 발생하여 반복 중단 
    
   

````python
for i in [1,2,3]:
    print(i)
    
print('for문을 해부해 본다면....')

it = iter([1,2,3])

print(next(it))
print(next(it))
print(next(it))
print(next(it)) 
````

## 연결리스트를 iterable로 만들기 

    연결리스트에서 iterator 클래스 생성 ( 반복자 클래스 ) 
    __iter__() 와 __next__() 
    
    이후 iterable 구현한다. ( 반복가능객체 )
    
```python

    class LinkedListIterator:
        
        def __init__( self, head: Node) -> None:
            self.current = head  머리 노드를 시작점으로 지정 
            
        def __iter__(self) -> LinkedListIterator: 
            return self 반드시 반복자를 리턴함 
        
        def __next__(self) -> Any:
            if self.current is None: 더 이상 반복되지 않으면 
                raise StopIteration 반복종료 
            else:
                data = self.current.data 현재 current 저장 
                self.current = self.current.next current 뒤로 이동 ( 반복시작 ) 
                return data 저장 
                
            
       [ 내부적으로 __iter__() 생성하기 ] 
        def __iter__(self) -> LinkedListIterator: 
            return LinkedListIterator(self.head)  
        
        내부에 __iter__() 생성함으로써 연결리스트 에서 
        for문을 실행할 수 있다. 
```

````python
class LinkedListIterator:
        
    def __init__( self, head: Node) -> None:
        self.current = head 
            
    def __iter__(self) -> LinkedListIterator: 
         return self 
        
    def __next__(self) -> Any:
        if self.current is None: 
            raise StopIteration  
        else:
            data = self.current.data 
            self.current = self.current.next 
            return data
    def __iter__(self) -> LinkedListIterator: 
        return LinkedListIterator(self.head)  
            
````
