## csv 파일 읽기 / csv.reader()

```python
import csv

f = open('test.csv', 'r') # 반드시 open 으로 시작하여 close 로 끝내야함  
rdr = csv.reader(f) # rdr = 읽거나 쓰기 

for line in rdr:
    print(line)
f.close() # close 로 끝 

## with 방식 , close 없음 
# import csv

# with open('test.csv', 'r') as f:
#     rdr = csv.reader(f)
#     for line in rdr:
#         print(line)
```

## csv 파일 쓰기 / csv.writer() , csv.writerow()

```python
import csv

f = open('output.csv', 'w', newline='', encoding='utf-8') # 'w' 생성, 만약 파일이름이 동일이름일 경우 덮어씌움 , 추가는 'a' 임 
# newline='' 빈문자열을 자동으로 넣어줘야 csv.writer()가 원활히 쓰기가능
# encoding='utf-8' , OS 환경에 따라 다름 

wr = csv.writer(f)
wr.writerow([1, "Alice", True]) # csv 내용물
wr.writerow([2, "Bob", False]) # csv 내용물 
wr.writerow([1, 'hyo', True ])
f.close()

## with 방식 

# import csv
# with open('output.csv', 'w', newline='', encoding='utf-8') as f:
#     wr = csv.writer(f)
#     wr.writerow([1, "Alice", True])
#     wr.writerow([2, "Bob", False])
```

## sheet 생성 및 불러오기 , 이름 변경 

```python
wb = openpyxl.Workbook()
sheet = wb.active

# sheet 생성
sheet2 = wb.create_sheet('2nd_sheet')
# sheet 불러오기 
sheet2 = wb['2nd_sheet']
# sheet 이름 바꾸기 
sheet2.title = 'data_sheet'


```

## cell 접근 ( 쓰기 , 읽기 ) 

```python
import openpyxl

wb = openpyxl.Workbook()
sheet = wb.active
 
sheet['A1'] = '이름' # A1 시트 위치 에 '이름' 
sheet.cell(row=2, column=1).value = '홍길동'

sheet.cell(row=1, column=2).value = '학번'
sheet.cell(row=2, column=2).value = 2012345

# 시트 맨 마지막에 데이터 (배트맨,조커) 넣기 
# 맨마지막 이란 ? = 데이터가 맨 마지막에 저장된 위치 ( = 홍길동의 위치 )
sheet.append(['배트맨', 2011111])
sheet.append(['조커', 2011112])

wb.save('test2.xlsx')
```
