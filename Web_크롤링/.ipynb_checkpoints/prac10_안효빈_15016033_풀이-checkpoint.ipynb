# 실습1

아래 네이버 파파고 번역 OpenAPI 문서를 참조하여 한글을 영어로 번역하시오.

https://developers.naver.com/docs/papago/papago-nmt-overview.md


- API를 요청할 때 다음 예와 같이 HTTP 요청 헤더에 클라이언트 아이디와 클라이언트 시크릿을 추가해야 합니다.

```
POST /v1/papago/n2mt HTTP/1.1
HOST: openapi.naver.com
User-Agent: curl/7.49.1
Accept: */*
Content-Type: application/x-www-form-urlencoded; charset=UTF-8
X-Naver-Client-Id: {애플리케이션 등록 시 발급받은 클라이언트 아이디 값}
X-Naver-Client-Secret: {애플리케이션 등록 시 발급받은 클라이언트 시크릿 값}
Content-Length: 51
```
- 응답 예
```json
{
    "message": {
        "@type": "response",
        "@service": "naverservice.nmt.proxy",
        "@version": "1.0.0",
        "result": {
            "srcLangType":"ko",
            "tarLangType":"en",
            "translatedText": "tea"
        }
    }
}
```
***결과***
```
Are you eating properly?
```


````python
import requests as rq

client_id = "AOQj_nTrjKrXqDTgfWfX" # 개인 ID
client_secret = "PXg6pfrep9" # 개인 PW 
org_text = '오늘은 2021년 6월 8일 입니다.' # 번역할 텍스트 

api_url = "https://openapi.naver.com/v1/papago/n2mt" # 요청 URL , Open API 제공  


# 요청 데이터 만들기 
data = {
    'source':'ko',
    'target':'en',
    'text':org_text.encode('utf-8') # utf-8으로 텍스트를 인코딩 , 한글이면 무조건 인코딩 필수 
}

# 요청 헤드 만들기 

headers = {
    "Content-Type": "application/x-www-form-urlencoded; charset=UTF-8",
    "X-Naver-Client-Id":client_id,
    "X-Naver-Client-Secret":client_secret
}
 
# post 방식 사용 , data 와 head 를 실어서 사용 
res = rq.post(api_url, data=data, headers=headers) # HTTP post 요청

# 받은 json 데이터를 사전형식으로 변환
res_data = res.json() 

# 요청에 성공했다면 , message 밑에 result 밑에 translatedText를 print ( 3가지 전부 key 임 )
if res.status_code == 200:
    print(res_data['message']['result']['translatedText'])
else:
    print ("Error Code:", res.status_code) # 오류시 상태코드 출력 
````

````python
import requests
from urllib.parse import urlencode

client_id = 'AOQj_nTrjKrXqDTgfWfX'
client_secret = 'PXg6pfrep9'
search_text = "갤럭시 핏2"

# get 메소드를 통해 XML,JSON 유형 중 JSON 출력 형태 선택 
api_url = 'https://openapi.naver.com/v1/search/shop.json'

# 요청 변수 
# query : 인코딩 , display : 디스플레이 갯수 , start : 검색 시작 위치 , sort : 정렬옵션 'sim' 
data = {
    'query':search_text.encode('utf-8'),
    'display':10,
    'start':1,
    'sort':'sim'
}

# 요청 헤더 , request 패키지가 모두 자동변수 이기에 딱히 설정 X , [6.변수-요청] 참고 
headers = {
    "X-Naver-Client-Id":client_id, 
    "X-Naver-Client-Secret":client_secret
}

request_url = api_url+"?"+urlencode(data) # urlencode() 함수는 사전을 쿼리스트링으로 바꾸어 준다.
print(urlencode(data)) # 쿼리스트링(데이터) , 해당 url 의 쿼리스트링이 data 이기 떄문 ? https://search.shopping.naver.com/gate.nhn 까진 모두 동일 ? 부터가 다르기 떄문에 쿼리스트링 = 크롤링 값 
print(request_url)



# 메소드가 post가 아닌 get 
res = requests.get(request_url, headers=headers) # header 만 넣어서 보여주기 

# 출력결과에 다양한 JSON key 값이 있다. 그 중 {title : , Lprice : , link : ] 데이터를 print 
if res.status_code == 200:
    res_data = res.json() # json -> dict
    for index, item in enumerate(res_data['items']):
        print (index + 1, item['title'], item['lprice'], item['link'])
else:
    print ("Error Code:", res.status_code)
````
